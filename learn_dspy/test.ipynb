```python
from langchain_community.tools import DuckDuckGoSearchRun

search = DuckDuckGoSearchRun()

search.invoke("Galgotias dog")
```

```python
def search_web(query):
    return search.invoke(query)
```

```python
import os
import dspy

import mlflow
mlflow.autolog()
mlflow.set_tracking_uri("http://127.0.0.1:5000")
mlflow.set_experiment("Reflection")

lm = dspy.LM(
    model="groq/openai/gpt-oss-20b",
    api_key=os.environ.get("GROQ_API_KEY"),
    base_url="https://api.groq.com/openai/v1",
)

dspy.configure(lm=lm)
```

```python
class SarcasmGenerator(dspy.Signature):
    """
    Generate a sarcasm based on some input text/idea
    """
    idea = dspy.InputField(description="An idea/thought to generate sarcasm on top of")
    thought = dspy.OutputField(description="Think of various comical and funny things about the idea...")
    sarcasm = dspy.OutputField(description="Sarcasm, HILARIOUS")
    concise_tight_sarcastic_joke = dspy.OutputField(description="Final take, HILARIOUS, FUNNY, TIGHT, 3 line sarcastic joke!")
```

```python
generate_sarcasm = dspy.ReAct(signature=SarcasmGenerator,
                     tools=[search_web], max_iters=1)

pred = generate_sarcasm(idea="Galgotia Dog")
```
