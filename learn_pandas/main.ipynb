### Sorting and Aggregation

```python
data = {
    "name": ["Arun", "Varun", "Tarun", "Karun", "Narun"],
    "age": [28, 34, 22, 34, 28],
    "salary": [27943, 78926, 89047, 16738, 37892],
}

df = pd.DataFrame(data)
```

```python
# Sort by age
df.sort_values(by="age", ascending=True, inplace=True)
df
```

```python
# Sort by age and salary
df.sort_values(by=["salary", "age"], ascending=[True, True], inplace=True)
df
```

```python
# Grouping by age
salary_of_ages = df.groupby(by="age")["salary"].sum()
salary_of_ages
```

```python
# Grouping by age
salary_of_ages_names = df.groupby(by=["age", "name"])["salary"].sum()
salary_of_ages_names
```

### Merging and Joining

```python
data1 = {
    "cust_id": [1,2,3,4, 5],
    "name": ["a", "b", "c", "d", "e"]
}

data2 = {
    "cust_id": [1,2,3,4],
    "purchased_item": [234, 345, 135, 354] 
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# df = pd.merge(df1, df2, on="column_name", how="type_of_join")
df = pd.merge(df1, df2, on="cust_id", how="left")
df
```

 ### Concatenation

```python
df_region1 = pd.DataFrame({
    "cust_id": [1,2],
    "name": ["ayush", "priya"]
})

df_region2 = pd.DataFrame({
    "cust_id": [3,4],
    "name": ["gourish", "yogesh"]
})

df = pd.concat([df_region1, df_region2], axis=0, ignore_index=True)
df
```

```python
df = pd.concat([df_region1, df_region2], axis=1, ignore_index=False)
df
```
