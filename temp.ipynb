```python
from ollama import chat
from ollama import ChatResponse

messages = [{'role': 'user', 'content': 'Why is the sky blue?'}]
stream = chat(
    model='llama3.2:latest',
    messages=[{'role': 'user', 'content': 'Why is the sky blue?'}],
    stream=True,
)

response = ""
for chunk in stream:
    response += chunk['message']['content']
    print(chunk['message']['content'], end='', flush=True)
```
