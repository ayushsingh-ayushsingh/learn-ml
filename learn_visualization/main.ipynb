```python
data = pd.read_csv("./final_vg.csv")

data.head()
```

```python
# Categorical columns
unique_categories = data["Genre"].unique()
unique_categories
```

```python
category_counts = data["Genre"].value_counts()
category_counts
```

```python
# Top 5 genres from the data
category_counts[0:5]
```

### Bar charts are used to represent the counts

```python
# Bar charts
x_bar = category_counts.index
x_bar
```

```python
y_bar = category_counts.values
y_bar
```

```python
plt.bar(x_bar, y_bar)
```

```python
plt.figure(figsize=(12, 8))
plt.bar(x_bar, y_bar)
```

```python
plt.figure(figsize=(12, 8))
plt.xticks(rotation=30, fontsize=12)
plt.bar(x_bar, y_bar, width=0.4, color="#ce3300")
plt.xlabel("Video Game Genre", fontsize=14)
plt.ylabel("Number of games", fontsize=14)
plt.title("Genre analysis", fontsize=24)

plt.show()
```

```python
# Using seaborn
sns.countplot(x="Genre",
              data=data,
              order=data["Genre"].value_counts().index,
              color="#1db868",
              width=0.6,
              )
plt.xticks(rotation=90)
plt.show()
```

### Pie charts are used to visualize proportions

```python
sales_data = data[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]]
regional_sales = sales_data.T.sum(axis="columns")
pd.DataFrame(regional_sales)
```

```python
total_sales = sales_data.sum().sum()
print(total_sales)
```

```python
plt.pie(x=regional_sales,
        labels=regional_sales.index,
        startangle=90,
        explode=[0.1] + [0] * (len(regional_sales) - 1),
        autopct="%1.1f%%",
        )
plt.title("Total Sales Across Regions")
plt.show()
```
